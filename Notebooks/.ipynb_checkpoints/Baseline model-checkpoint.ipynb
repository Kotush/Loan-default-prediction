```python
# Read CSV file
df = pd.read_csv(dataset_path, encoding="utf-8")
df.head()
```

__Recursive feature elimination__

```python

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

# Step 1: Define X and y
X = df.drop(columns=['loan_default'])  # Drop target column 'loan_default'
y = df['loan_default']  # Target column 'loan_default'

# Drop the unwanted features from X due to high correlation with last_fico_range_high and loan default
X = X.drop(columns=['last_fico_range_low', 'loan_status'])

# Step 2: Identify categorical columns and encode them
categorical_columns = X.select_dtypes(include=['object']).columns

# Apply Label Encoding to categorical columns
label_encoder = LabelEncoder()
for col in categorical_columns:
    X[col] = label_encoder.fit_transform(X[col])

# Step 3: Select numeric and boolean columns (ensure we're only working with numeric and boolean columns for RFE)
numeric_bool_columns = X.select_dtypes(include=['number', 'bool']).columns

# Step 4: Replace NaN or inf in numeric and boolean columns (if any)
X[numeric_bool_columns] = X[numeric_bool_columns].replace([np.inf, -np.inf], np.nan)

# Step 5: Impute missing values in these columns with the mean
imputer = SimpleImputer(strategy='mean')
X[numeric_bool_columns] = imputer.fit_transform(X[numeric_bool_columns])

# Step 6: Convert boolean columns to numeric (True -> 1, False -> 0)
X[numeric_bool_columns] = X[numeric_bool_columns].astype(int)

# Step 7: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 8: Initialize RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Step 9: Apply Recursive Feature Elimination (RFE) only to numeric features
rfe = RFE(estimator=rf, n_features_to_select=20, step=1, verbose=3)
rfe.fit(X_train, y_train)

# Step 10: Check selected features and their rankings
selected_features = X_train.columns[rfe.support_]  # Features selected by RFE
print(f"Selected features: {selected_features}")
print(f"Ranking of features: {rfe.ranking_}")  # 1 means selected, higher numbers indicate less importance



```

__Importance of top features__

```python
# Fit RandomForestClassifier directly to get feature importances
rf.fit(X_train, y_train)
```

```python
# Extract feature importances from the fitted RFE model (from the internal estimator)
selected_importances = rfe.estimator_.feature_importances_
```

__View Feature Importances in Sorted Format:__

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with feature names and their importance
feature_imp_df = pd.DataFrame({
    'Feature': selected_features,
    'Importance': selected_importances
}).sort_values(by='Importance', ascending=True)

# Plot the top selected features
plt.figure(figsize=(10, 6))
plt.barh(feature_imp_df['Feature'], feature_imp_df['Importance'])
plt.xlabel('Importance')
plt.title('Top Feature Importances (Random Forest + RFE)')
plt.tight_layout()
plt.show()
```

__Preparing data for Logistic Regression__

__Training a Logistic Regression model after feature selection and imputation__

```python
# STEP 1: Rebuild DataFrame from imputed NumPy array

full_feature_names = df.drop(columns='loan_default').columns  # All original feature names
X = pd.DataFrame(X, columns=full_feature_names)  # Re-assign proper column names

# STEP 2: Select only top 20 features selected by RFE
X_selected = X[selected_features]  # selected_features came from your RFE output

# STEP 3: Impute missing values for selected features (if not already imputed)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_selected = pd.DataFrame(imputer.fit_transform(X_selected), columns=selected_features)

# STEP 4: Split into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# STEP 5: Train Logistic Regression model
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# STEP 6: Make predictions
y_pred = logreg.predict(X_test)

# STEP 7: Evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))



```

__Threshold tuning__

```python
threshold = 0.6
y_pred = (logreg.predict_proba(X_test)[:, 1] > threshold).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

__Plotting__

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Get predicted probabilities for the positive class
y_proba = logreg.predict_proba(X_test)[:, 1]

# Compute ROC curve and ROC AUC
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2, label='Random baseline')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.tight_layout()
plt.show()
```

The True Positive Rate (Recall) is high even when the False Positive Rate is low, especially in the left part of the graph â€” this is ideal for imbalanced problems like fraud or loan default detection.

# Summary of Baseline Model Performance (Logistic Regression after Feature Selection)


We built a baseline classification model to predict loan default using Logistic Regression.

__Key steps:__
- Cleaned and preprocessed the data (missing values, encoding, scaling).
- Selected the top 20 most important features using Recursive Feature Elimination (RFE) with a Random Forest.
- Trained a Logistic Regression model using only the selected features.
- Evaluated performance using classification metrics and confusion matrix.

__Model results (test set of 20,000 samples):__
- Accuracy: 93%
- Precision (Default = 1): 77%
- Recall (Default = 1): 57%
- F1-score (Default = 1): 65%
- Weighted Avg F1-score: 0.92

__Note: The model reached the maximum number of iterations and did not fully converge. This can be addressed by increasing `max_iter`.__

__Insights:__
- The model performs very well in identifying non-defaults (class 0).
- However, recall for defaults (class 1) is moderate, meaning some defaults are still missed.
- Top predictive features include FICO scores, recent credit activity, loan amount, and utilization ratios.

__Possible Next steps:__
- Try `class_weight='balanced'` to improve recall on defaults.
- Experiment with different models like Random Forest or XGBoost.
- Tune the classification threshold for business-specific risk tolerance.
- Use SHAP or permutation importance for better interpretability.
